```python
import pandas as pd
merge = pd.read_csv("merge.csv")
merge
```

```python
df = pd.DataFrame(columns=('time', 'activity', 'percent_0', 'percent_1', 'percent_2', 'percent_3'))
time = []
for x in range (0, 60*24, 5):
    for i in range(0,10):
        time.append(x)
df['time'] = time
df  
```

```python
total_acts = []
acts = ['sleep', 'socializing', 'eating', 'household', 'shopping', 'family', 'work', 'education', 'sports', 'entertainment']
for x in range (0, 60*24, 5):
    total_acts.extend(acts)
df['activity'] = total_acts
df
```

```python
def f(row):
    row['percent_0'] = 0
    row['percent_1'] = 0
    row['percent_2'] = 0
    row['percent_3'] = 0
    return row

df = df.apply(f, axis=1)
```

```python
incomes = [0,0,0,0]
#for num in merge['income']:
#    incomes[num] += 1
for i in range (0,4):
    incomes[i] = merge[merge['income'] == i]['id'].nunique()
incomes
```

```python
def num(col, act, time):
    ff = merge[(merge['category'] == act) & (merge['income'] == col)]
    count = ff.shape[0]
    new = ff.drop_duplicates(subset ="id", inplace = False)   
    
    total = 0
    time = int(time)
    
    # nominal case - time is >= start, < end
    inRange = new[(new['start'] <= time) & (new['end'] > time)]
    total = inRange.shape[0]
    
    # edge case where end wraps around so end < start
    edge = new[new['end'] < new['start']]
    edge['end'] += 1440
    # case 1 example: start = 1260, end = 1920, time = 1400
    edge1 = edge[(edge['start'] <= time) & (edge['end'] > time)]
    # case 2 example: start = 1260, end = 1920, time = 5
    edge2 = edge[(edge['start'] <= time + 1440) & (edge['end'] > time + 1440)]
    total += edge1.shape[0] + edge2.shape[0]
    
    return total / count

def f(row):
    row['percent_0'] = num(0, row['activity'], row['time'])# / incomes[0]
    row['percent_1'] = num(1, row['activity'], row['time'])# / incomes[1]
    row['percent_2'] = num(2, row['activity'], row['time'])# / incomes[2]
    row['percent_3'] = num(3, row['activity'], row['time'])# / incomes[3]
    return row

df = df.apply(f, axis=1)
df
```

```python
df.to_csv("aggregate-all.csv")
```
