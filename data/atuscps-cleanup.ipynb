```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 500)
df = pd.read_csv("atuscps.csv")
df.head()
```

```python
df1 = df.filter(['tucaseid', 'tulineno', 'hefaminc', 'hufaminc', 'prempnot'])
df1
```

```python
df1['income'] = df1['hefaminc'] + df1['hufaminc'] + 1
df2 = df1.drop(columns=['hefaminc', 'hufaminc'])
df2 = df2.rename(columns={'income': 'hefaminc'})
df2
```

```python
df2 = df2[df2['hefaminc'] > 0]
df2
```
