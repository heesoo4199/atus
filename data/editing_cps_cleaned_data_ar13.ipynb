```python
df = pd.read_csv("atuscps-clean.csv");
```

```python
# drop the unnecessary columns.
df.drop("Unnamed: 0", axis=1, inplace=True)
df.drop("tulineno", axis=1, inplace=True)
df.drop("prempnot", axis=1, inplace=True)
```

```python
# drop rows with duplicate ids.
df.sort_values("tucaseid", inplace = True)
df.drop_duplicates(subset ="tucaseid", inplace = True)
df
```

```python
# 1 - 11  = 0
# 12 - 14 = 1
# 15      = 2
# 16      = 3

# Replace all values between 1 - 11 to 0.
df.replace(to_replace=1, value=0, inplace=True)
df.replace(to_replace=2, value=0, inplace=True)
df.replace(to_replace=3, value=0, inplace=True)
df.replace(to_replace=4, value=0, inplace=True)
df.replace(to_replace=5, value=0, inplace=True)
df.replace(to_replace=6, value=0, inplace=True)
df.replace(to_replace=7, value=0, inplace=True)
df.replace(to_replace=8, value=0, inplace=True)
df.replace(to_replace=9, value=0, inplace=True)
df.replace(to_replace=10, value=0, inplace=True)
df.replace(to_replace=11, value=0, inplace=True)

# Replace all values from 12 - 14 to 1.
df.replace(to_replace=12, value=1, inplace=True)
df.replace(to_replace=13, value=1, inplace=True)
df.replace(to_replace=14, value=1, inplace=True)

# Replace all values from 15 to 2.
df.replace(to_replace=15, value=2, inplace=True)

# Replace all values from 16 to 3.
df.replace(to_replace=16, value=3, inplace=True)
```
