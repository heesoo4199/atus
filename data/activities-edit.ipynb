```python
import pandas as pd
df = pd.read_csv("atusact-clean.csv")
df.head()
```

```python
# Only keep rows where activity belongs to category we care about
"""Categories:
* Socializing
    * 1201
    * 1202
    * 1299
* Eating and Drinking
    * 11
* Household Activities
    * 02
* Consumer
    * 07
* Family Care
    * 03
* Work
    * 05
* Education
    * 06
* Sleeping
    * 0101
* TV/Games
    * 1203
    * 1204
* Sports/Exercise
    * 1301
    * 1399
"""
df = df[(df['trtier2p'] == 1201) | (df['trtier2p'] == 1202) | (df['trtier2p'] == 1203) | (df['trtier2p'] == 1204) | (df['trtier2p'] == 1299) | (df['trtier2p'] == 1301) | (df['trtier2p'] == 1399) | (df['trtier2p'] == 101) | (df['trtier1p'] == 11) | (df['trtier1p'] == 2) | (df['trtier1p'] == 3) | (df['trtier1p'] == 5) | (df['trtier1p'] == 6) | (df['trtier1p'] == 7)]
df
```

```python
# further drop unused columns
df = df.drop(columns=['tuactdur24', 'trcodep', 'tucumdur', 'tucumdur24', 'Unnamed: 0'])
df
```

```python
# create new column with activitiy category (string)
def f(row):
    if row['trtier2p'] == 101:
        val = 'sleep'
    elif row['trtier1p'] == 2:
        val = 'household'
    elif row['trtier1p'] == 3:
        val = 'family'
    elif row['trtier1p'] == 5:
        val = 'work'
    elif row['trtier1p'] == 6:
        val = 'education'
    elif row['trtier1p'] == 7:
        val = 'shopping'
    elif row['trtier1p'] == 11:
        val = 'eating'
    elif (row['trtier2p'] == 1201) or (row['trtier2p'] == 1202) or (row['trtier2p'] == 1299):
        val = 'socializing'
    elif (row['trtier2p'] == 1203) or (row['trtier2p'] == 1204):
        val = 'entertainment'
    #elif row['trtier2p'] > 1300:
    else:
        val = 'sports'
    return val

df['category'] = df.apply(f, axis=1)
df
```

```python
df = df.drop(columns=['tuactivity_n', 'trtier1p', 'trtier2p'])
df
```

```python
cps = pd.read_csv("atuscps-cleaned-again.csv")
cps = cps.drop(columns=['Unnamed: 0'])
cps
```

```python
merge = pd.merge(df, cps, on='tucaseid')
merge
```

```python
merge.columns = ['id','stime','etime', 'duration','category','income']
merge
```

```python
time = pd.DatetimeIndex(merge['stime'])
merge['start'] = time.hour * 60 + time.minute

time = pd.DatetimeIndex(merge['etime'])
merge['end'] = time.hour * 60 + time.minute

merge
```

```python
merge = merge.drop(columns=['stime', 'etime'])
merge
```

```python
merge = merge.drop(columns=['duration'])
merge
```

```python
merge.to_csv("merge.csv")
```
